# Model design
## Identify which team would be the champion of the Colombian professional soccer league, according to the behavior of the results of the last leagues.

Colombian League - 20 soccer teams

Series of 1-1 duels over 19 or 20 dates - All against all.

Normally there are 19 dates. One more is added during the classic round.

Final quadrangulars
2 groups of 4 teams, and the first of each quadrangular advances to the final.

6 games.

In some tournaments there are play-offs - quarter-finals, semi-finals -

Final - 2 games of the same teams.

```python
import pandas as pd
import numpy as np
```

```python
rute = "https://es.wikipedia.org/wiki/Torneo_Apertura_2024_(Colombia)"
# Read the HTML content and convert it to a Pandas DataFrame,
# added the web sections in an array
# So useful for a fast exploration of the data
tables = pd.read_html(rute)
```

## First example for data frame

```python
tables[8]
```

```python
df_positions = tables[8]
df_positions
```

```python
# Delete the classification column

try:
    df_positions = df_positions.drop(columns=['Clasificación'])
except Exception as e:
    print(f"Error delete column: {e}")

df_positions
```

```python
df_positions.info()
```

```python
# Replace the +  of the gol difference
df_positions['Dif.'] = df_positions['Dif.'].str.replace('+', '')
df_positions
```

```python
# Replace the - of the column Dif
df_positions['Dif.'] = df_positions['Dif.'].str.replace('−','-')
df_positions
```

```python
# Convert the column data Dif. to integer
df_positions['Dif.'] = df_positions['Dif.'].astype(int)
df_positions.info()
```

```python
df_positions
```

## Second example for Data Frame 

```python
df_pos2 = tables[8]
df_pos2
```

```python
## Drop "Clasificación" column

try: 
    df_pos2 = df_pos2.drop(columns=['Clasificación'])
    df_pos2['Dif.'] = df_pos2['Dif.'].str.replace('−','-')
    df_pos2['Dif.'] = df_pos2['Dif.'].str.replace('^\d-','',regex=True)
    df_pos2['Dif.'] = df_pos2['Dif.'].astype(int)
except Exception as e:
    print(e)
df_pos2.head(5)
```

```python
# Rename the columns
df_pos2.rename(columns = {'Pos.':'Position','Pts.':'Score','Dif.':'Dif. Gol'}, inplace=True)
df_pos2
```

```python
# Export to csv file
df_pos2.to_csv('../data/df_pos2.csv', index=False)
```
