### Clear and extract of final quadrangulars 

### Quadrangular A

```python
# Quadrangular A
tables[33]

# Assign the table to a new df
quadrangularA = tables[33]

# Delete the column of level 0
try:
    quadrangularA.columns = quadrangularA.columns.droplevel(0)
except:
    pass

# Delete the unnecessary columns
quadrangularA.drop(columns = quadrangularA.columns[0], inplace = True)
quadrangularA.drop(columns = quadrangularA.columns[3:], inplace = True)
quadrangularA
```

```python
# Separate the "Resultado" column
quadrangularA[['Local goals','Visit goals']] = quadrangularA['Resultado'].str.split(':', expand=True)
quadrangularA
```

```python
# Delete the column "Resultado"
quadrangularA.drop('Resultado',axis=1,inplace=True)
```

```python
quadrangularA.dtypes
```

```python
# Try convert the column goals to numeric

try:
    quadrangularA['Local goals'] = quadrangularA['Local goals'].astype(int)
except Exception as e:
    print(f"Error converting the column 'Local goals': {e}")

try:
    quadrangularA['Visit goals'] = quadrangularA['Visit goals'].astype(int)
except Exception as e:
    print(f"Error converting the column 'Visitant goals': {e}")

quadrangularA.dtypes
```
