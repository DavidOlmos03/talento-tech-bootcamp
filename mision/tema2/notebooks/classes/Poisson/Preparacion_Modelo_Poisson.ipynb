# Identifiquemos la fortaleza o poder de cada equipo.

```python
#dividimos el dataframe para identificar los goles de cada equipo
df_local = torneo_2024_1[['Local', 'Goles Local', 'Goles Visitante']]
df_visitante = torneo_2024_1[['Visitante', 'Goles Local', 'Goles Visitante']]
```

```python
#observemos que nos quedó
df_local.head()
df_visitante.head()

print(f"{df_local.loc[0]} \n {df_visitante.loc[0]}")
```

```python
#Renombramos las columnas
df_local.rename(columns={'Local':'Equipo','Goles Local':'Goles_Favor', 'Goles Visitante':'Goles_Contra'}, inplace=True)
df_visitante.rename(columns={'Visitante':'Equipo', 'Goles Local':'Goles_Contra', 'Goles Visitante': 'Goles_Favor'}, inplace=True)
```

```python
print(f"{df_local.loc[0]} \n {df_visitante.loc[0]}")
```

```python
#Concatenamos los df
pd.concat([df_local, df_visitante], ignore_index=True)
```

```python
df_equipo_poder = pd.concat([df_local, df_visitante], ignore_index=True).groupby('Equipo').mean()
df_equipo_poder
```
