```python
import pandas as pd
import numpy as np
```

```python
ruta = "https://es.wikipedia.org/wiki/Torneo_Apertura_2024_(Colombia)"
tablas = pd.read_html(ruta)
```

```python
tablas[8]
```

```python
df_posiciones = tablas[8]
df_posiciones
```

```python
#eliminar la columna clasificación

try:
    df_posiciones = df_posiciones.drop(columns=['Clasificación'])
except Exception as e:
    print(e)
df_posiciones
```

```python
df_posiciones.info()
```

```python
#Vamos a reemplazar el + de la columna diferencia de gol
df_posiciones['Dif.'] = df_posiciones['Dif.'].str.replace('+', '')
df_posiciones
```

```python
#Vamos a reemplazar el - de la columna diferencia de gol
df_posiciones['Dif.'] = df_posiciones['Dif.'].str.replace('−', '-')
df_posiciones
```

```python
#convertir la columna dif. en entero
df_posiciones['Dif.'] = df_posiciones['Dif.'].astype(int)
df_posiciones.info()
```

```python
df_posiciones
```

```python
df_pos2 = tablas[8]
df_pos2
```

```python
#quitarle la columna clasificación.
try:
    df_pos2 = df_pos2.drop(columns=['Clasificación'])
    df_pos2['Dif.'] = df_pos2['Dif.'].str.replace('−','-')
    df_pos2['Dif.'] = df_pos2['Dif.'].str.replace('^\d-','',regex=True)
    df_pos2['Dif.'] = df_pos2['Dif.'].astype(int)
except Exception as e:
    print(e)
df_pos2.head(1)
```

```python
#renombrar las columnas
df_pos2.rename(columns={'Pos.':'Posición','Pts.':'Puntos','Dif.':'Dif. Gol'}, inplace=True)
df_pos2
```

```python
#exportamos a un archivo .csv
df_pos2.to_csv('posiciones_liga_2024_1.csv', index=False)
```
