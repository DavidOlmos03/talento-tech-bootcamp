```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/CART_MODEL/Datos_Pacientes.csv')
df
```

```python
#Normalicemos la edad
scaler = MinMaxScaler()
df['Edad'] = scaler.fit_transform(df[['Edad']])
df
```

```python
#limpiamos la columna enfermedad
df['Enfermedad'] = df['Enfermedad'].str.strip()
df
```

```python
#normalicemos la enfermedad
df['Enfermedad'] = df['Enfermedad'].map({'SI': 1, 'NO': 0})
df
```
