```python
lista = []
for i in range(10,29,1):
    df = tablas[i]
    try:
        df.columns = df.columns.droplevel(0)     #Se ejecuta la primera vez, luego lo comentamos porque ya quitamos la columna de nivel 0
    except:
        pass
    lista.append(df)
    #print(df)
#lista
todosContraTodos = pd.DataFrame()
for i in range(len(lista)):
    todosContraTodos = pd.concat([todosContraTodos, lista[i]], axis=0, ignore_index=True)
todosContraTodos
```

# Limpieza y extracción en los cuadrangulares finales

## Cuadrangular A

```python
#Cuadrangular A
tablas[33]
#asignar la tabla a un nuevo df
cuadrangularA = tablas[33]
#quitamos la columna de nivel 0
try:
    cuadrangularA.columns = cuadrangularA.columns.droplevel(0)
except:
    pass
#quitamos las columnas que no necesitamos
cuadrangularA.drop(columns=cuadrangularA.columns[0], inplace=True)
cuadrangularA.drop(columns=cuadrangularA.columns[3:], inplace=True)
cuadrangularA
```

```python
#separamos la columna de resultados
cuadrangularA[['Goles Local', 'Goles Visitante']] = cuadrangularA['Resultado'].str.split(':', expand=True)
cuadrangularA
```

```python
#quitamos la columna resultado
cuadrangularA.drop('Resultado', axis=1, inplace=True)
```

```python
cuadrangularA.dtypes
```

```python
#intentemos convertir las columnas de los goles en números
try:
    cuadrangularA['Goles Local'] = cuadrangularA['Goles Local'].astype(int)
except Exception as e:
    print(f"Error al convertir la columna 'Goles Local': {e}")

try:
    cuadrangularA['Goles Visitante'] = cuadrangularA['Goles Visitante'].astype(int)
except Exception as e:
    print(f"Error al convertir la columna 'Goles Visitante': {e}")

cuadrangularA.dtypes
```
