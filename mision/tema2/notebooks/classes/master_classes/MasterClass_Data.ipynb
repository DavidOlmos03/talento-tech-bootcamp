# Funciones básicas de las series

```python
numeros = pd.Series([1,2,3,4,5,6,7,8,9])
numeros.sum()  #suma los elementos de la serie
numeros.max() #máximo
numeros.min() #mínimo
numeros.std()  #desviación estandar
numeros.describe() #resumen estadístico
numeros
```

```python
dic_materias = {'Matemáticas':5.0,
                'Física':4.8,
                'Química':3.5,
                'Programación':2.8}
df_materias = pd.Series(dic_materias)
df_materias.describe()

#identificar si ganó la materia
df_gano = pd.Series(df_materias[df_materias>3])  #de esta manera nos muestra la nota de las materias ganadas
df_gano = pd.Series(df_materias>3) #Responde a la pregunta en forma de booleano
df_materias.sort_values(ascending=True)
df_materias.sort_values(ascending=False)
```

# Mezclar Dataframes

```python
datos1 = pd.DataFrame({'MOTOS':['Boxer','Invicta','GS500'],
                       'COLOR':['Negro','Blanca','Azul'],
                       })
datos1
```

```python
datos2 = pd.DataFrame({'MOTOS':['Pulsar','Invicta','GS500'],
                       'MODELO':[2024, 2021, 2016],
                       })
datos2
```

```python
datos_final = pd.merge(datos1, datos2, on='MOTOS', how='right')
datos_final

#inner es el que está por defecto y las une por columnas
#outer es el que está por defecto y las une por filas
#left combina filas y luego trae todos los datos del primer df
#right combina filas y luego trae todos los datos del segundo df
```

```python
datos_final2 = pd.merge(datos1, datos2, on='MOTOS', how="inner")
datos_final2
```
