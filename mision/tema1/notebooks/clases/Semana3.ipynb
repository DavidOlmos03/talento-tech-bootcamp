Importar las librerías necesarias para trabajar

```python
from sqlalchemy import create_engine, text
import pandas as pd
import sqlite3
```

Configuración de la base de datos

```python
#cadena de conexión
connection_string = 'sqlite:///ejemplo_fts.db'
#motor de conexión
connection = create_engine(connection_string)
```

DEfinir nuestra función de ejecución de los query

```python
from ast import Return
def run_query(query: str):
    with connection.connect() as conn:
        #para modificar
        if query.strip().lower().startswith(('insert','update','delete','create')):
            conn.execute(text(query))
            conn.commit()
        #para consultar
        else:
            res = conn.execute(text(query))
            result = pd.DataFrame(res.fetchall(), columns=res.keys())
            return result
```

Creación de la tabla para el ejemplo

```python
#creación de tabla
create_table_query = """
                    CREATE TABLE productos (
                        product_id INTEGER PRIMARY KEY,
                        product_name TEXT,
                        product_price INTEGER
                    );
                    """
create_table = run_query(create_table_query)
```

**CONFIGURAR EL FTS**

Creamos el catálogo de texto completo

```python
#creación de catálogo
create_catalog_query = """
                        CREATE VIRTUAL TABLE catalogo_productos USING fts5(
                            product_id,
                            product_name,
                            product_price
                        )
                        """

create_catalog = run_query(create_catalog_query)
```

configuramos el catálogo

```python
insert_data_query_virtual = """
                            INSERT INTO catalogo_productos(
                                product_id,
                                product_name,
                                product_price)
                            SELECT
                                product_id,
                                product_name,
                                product_price
                            FROM productos;
                            """
insert_data_virtual = run_query(insert_data_query_virtual)
```

intentemos mostrar la tabla virtual

```python
vista_query_virtual = "SELECT * FROM catalogo_productos "
vista_virtual = run_query(vista_query_virtual)
vista_virtual
```

Creamos el índice de texto completo

```python
#creamos el índice
create_fts_index_query = """
                            SELECT
                               t.product_id,
                               t.product_name,
                               t.product_price
                            FROM productos AS t
                            INNER JOIN
                                catalogo_productos AS v
                            ON
                                v.product_id = t.product_id
                            WHERE
                                catalogo_productos MATCH '15'
                            ORDER BY
                                bm25(catalogo_productos, 0, 10);

                        """
create_fts_index = run_query(create_fts_index_query)
create_fts_index
```
